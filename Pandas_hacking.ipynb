```python
import pandas as pd
```

```python
# initialize list of lists
# Rank,Title,Studio,Gross,Year
# 1,Avengers: Endgame,Buena Vista,"$2,796.30",2019
# 2,Avatar,Fox,"$2,789.70",2009

movies = [
    ['1', "Avengers", 'Century', "$2,796.30", 2019],
    ['2', "Avatar", 'Century', "$2,796.30", 2012],
    ['3', "Avatar 2", 'Buena Vista', "$2,796.30", 2012],
    ['4', "Avatar 3", 'Buena Vista', "$2,796.30", 2012],
    ['5', "Avatar 4", 'Buena Vista', "$2,796.30", 2012],
    ['6', "Avatar 5", 'Buena Vista', "$2,796.30", 2012],
    ['7', "Nemo", 'Pixar', "$2,796.30", 2012]
  ]

# Create the pandas DataFrame
movies = pd.DataFrame(movies, columns=['Rank', 'Title', 'Studio', 'Gross', 'Year'])
movies.set_index("Title")


# print dataframe.
movies
```

```python
# Print only first x rows
movies.head(2)
```

```python
# More fun operations
print("Number of rows ", len(movies))

# Top 2 rows
movies.head(2)

print("Size of the matrix\n", movies.shape)

print("Total number of cells\n", movies.size)

print("Element at index 2\n", movies.iloc[1])

```

```python
# Frequency count
movies["Studio"].value_counts()
```

```python
# top studios
movies["Studio"].value_counts().head(2)
```
